### Missing Values

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
print(features_with_na)
```

```python
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

### Since there are lot of missing values,we need to find the relationship between missing value and sales value

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['b','g'])
    plt.title(feature)
    plt.show()
    
```

## Numerical Variables

```python
# list of numerical variables
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
```

```python
print(numerical_feature)
print('number of numerical variables :',len(numerical_feature))
```

```python
dataset[numerical_feature].head()
```

### finding Temporal veriable(datetime variables)

```python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
dataset[year_feature].head()
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Sold Year')
plt.ylabel('Median House Price')
plt.title('HousePrice vs YearSold')
#plt.show()
```

```python
### here we will compare difference between all years and sales price
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        ##plt.title(feature 'vs SalePrice')
        plt.show()
```

```python
## variables are two types 1. discrete and 2.contineous
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('number of discrete features are:',len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
## find out relationship between discrete variable and salesprice
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
contineous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
print('number of contineous features are {}'.format(len(contineous_feature)))
```

```python
dataset[contineous_feature].head()
```

```python
## lets analyse contineous variable by maing histogram
for feature in contineous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
## we will be using logarithomic transformation
for feature in contineous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

### outliers


```python
for feature in contineous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        #plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

## catagorical feature

```python
catagorical_feature=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
dataset[catagorical_feature].head()
#print(len(catagorical_feature))
#dataset['Street']
```

```python
for feature in catagorical_feature:
    print('the feature name is {} and having cardinality number {}'.format(feature,len(dataset[feature].unique())))
```

```python
# relationship between catagorical feature and dependent variable
for feature in catagorical_feature:
    data=dataset.copy()
    #data.groupby(feature)['SalePrice'].median().plot.bar()
   # plt.xlabel(feature)
    #plt.ylabel('SalePrice')
    sns.barplot(x=feature,y='SalePrice',data=data,ci=False,estimator=np.median)
    plt.title(feature)
    plt.show()
```
